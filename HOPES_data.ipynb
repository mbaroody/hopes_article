```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
# Data
years = [2021, 2022, 2023, 2024]
patients_served = [456, 704, 712, 438]
student_volunteers = [108, 141, 398, 293]
physician_volunteers = [68, 69, 104, 114]

clinics = [
    'CCC', 'Dermatology', 'ENT', "Gyn/Women's Health",
    'Mental Health', 'Ophthalmology', 'Orthopedics', 'Primary Care',
    'Sports Medicine', 'Esperanza', 'Esperanza Peds', 'Life Counseling'
]

# Patient numbers per clinic (aligned with years)
clinic_data = {
    'CCC': [45, 43, 45, 49],
    'Dermatology': [22, 28, 46, 29],
    'ENT': [2, 6, 4, 6],
    "Gyn/Women's Health": [52, 79, 87, 51],
    'Mental Health': [13, 14, 41, 17],
    'Ophthalmology': [29, 33, 32, 23],
    'Orthopedics': [6, 25, 25, 9],
    'Primary Care': [71, 167, 189, 180],
    'Sports Medicine': [13, 53, 47, 18],
    'Esperanza': [113, 140, 0, 0],
    'Esperanza Peds': [72, 61, 0, 0],
    'Life Counseling': [13, 46, 0, 0]
}

locations = [
    'Norfolk', 'Chesapeake', 'Virginia Beach', 'Portsmouth', 'Hampton', 'Suffolk',
    'Franklin', 'Newport News', 'Poquoson', 'Williamsburg', 'Other VA city', 'Outside VA', 'Unknown'
]

geo_data = {
    2021: [421, 59, 88, 21, 1, 2, 0, 9, 0, 0, 5, 0, 24],
    2022: [397, 36, 107, 18, 11, 6, 0, 6, 0, 0, 9, 4, 39],
    2023: [397, 82, 82, 42, 23, 14, 0, 8, 0, 0, 16, 0, 22],
    2024: [254, 39, 82, 28, 3, 6, 0, 1, 0, 0, 7, 0, 11],
}

# Colors
primary_color = '#003156'
secondary_color = '#818a8f'
```

```python
# Plot for patients served
plt.figure(figsize=(8, 5))
bars = plt.bar(years, patients_served, color=primary_color)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height / 2, str(height),
             ha='center', va='bottom', fontsize=10, color=secondary_color)

# Titles and labels
plt.title('Number of patients served per year at HOPES Clinic',
          color=secondary_color, fontsize=14)
plt.xlabel('Year', color=secondary_color)
plt.ylabel('Patients Served', color=secondary_color)
plt.xticks(years, color=secondary_color)
plt.yticks(color=secondary_color)
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

# Remove border
plt.box(False)

plt.tight_layout()
# Save the figure
plt.savefig('hopes_patients_served.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Plot settings
bar_width = 0.6 / len(clinics)  # narrow bars to fit many clinics
x = np.arange(len(years))  # the label locations

# Create plot
plt.figure(figsize=(15, 8))

# Assign a unique color to each clinic automatically
colors = plt.cm.tab20(np.linspace(0, 1, len(clinics)))

# Plot each clinic
for idx, (clinic, counts) in enumerate(clinic_data.items()):
    bars = plt.bar(x + idx * bar_width, counts, width=bar_width, label=clinic, color=colors[idx])
    # Add value labels inside bars
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Only label non-zero bars
            plt.text(
                bar.get_x() + bar.get_width() / 2,   # x position
                height / 2,                          # y position: middle of bar
                str(int(height)),                    # label
                ha='center', va='center',             # center align both horizontally and vertically
                fontsize=7, color='white'             # small white font
            )

# Titles and labels
plt.title('Patients served per clinic by year', color='#818a8f', fontsize=16)
plt.xlabel('Year', color='#818a8f', fontsize=12)
plt.ylabel('Patients Served', color='#818a8f', fontsize=12)
plt.xticks(x + bar_width * (len(clinics)/2), years, color='#818a8f')
plt.yticks(color='#818a8f')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Remove border
plt.box(False)

plt.tight_layout()
plt.savefig('hopes_patients_served_by_clinic.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
cmap = cm.get_cmap('tab20', len(locations))
color_map = {loc: cmap(i) for i, loc in enumerate(locations)}

# Plot
fig, axs = plt.subplots(2, 2, figsize=(16, 10))
axs = axs.flatten()

def autopct_filter(pct):
    return f'{pct:.1f}%' if pct >= 3.5 else ''

for idx, year in enumerate(years):
    ax = axs[idx]
    data = geo_data[year]

    # Filter out zero entries
    filtered = [(loc, count) for loc, count in zip(locations, data) if count > 0]
    labels, values = zip(*filtered)
    colors = [color_map[loc] for loc in labels]

    wedges, texts, autotexts = ax.pie(
        values,
        colors=colors,
        startangle=140,
        autopct=autopct_filter,
        textprops={'fontsize': 8},
        pctdistance=0.7
    )

    ax.set_title(f'{year}', fontsize=14, color='#818a8f')

    # Add legend outside
    ax.legend(
        wedges, labels,
        title='Location',
        loc='center left',
        bbox_to_anchor=(1, 0.5),
        fontsize=8
    )

# Overall title and spacing
plt.suptitle('Geographic Distribution of Patients Served by Year', fontsize=16, color='#818a8f')
plt.tight_layout(rect=[0, 0, 0.9, 0.95])
plt.savefig('hopes_patients_served_by_region.png', dpi=300, bbox_inches='tight')
plt.show()
```
